```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def Threshold_1(img, l, r):
    rows, cols = img.shape
    newMatrix = [[] for i in range(rows)]
    for i in range(rows):
        for j in range(cols):
            newMatrix[i].append(Change_1(img[i, j], l, r))
    plt.subplot(2, 2, 3), plt.imshow(newMatrix, 'gray')
```

```python
def Change_1(val, l, r):
    if(val > l and val < r):
        return 255
    else:
        return 0
```

```python
img = cv.imread('thresh1.png', 0)
hist = cv.calcHist([img], [0], None, [256], [0, 256])
plt.subplot(2, 2, 1), plt.imshow(img, 'gray')
plt.subplot(2, 2, 2), plt.imshow(hist, 'gray'), plt.xlim([0, 256])
Threshold_1(img, 180, 230)
plt.show()
```
